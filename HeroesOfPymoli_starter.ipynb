```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
#find the unique players and ascend
```

```python
total_players = purchase_data["SN"].nunique()
total_players
```

```python
#to avoid scaler issues must put [] around defined variable
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Item_Count = purchase_data["Item ID"].nunique()
Item_Count

Average_Price = round(purchase_data["Price"].mean(),2)
Average_Price

Number_Purchases = purchase_data["Purchase ID"].count()
Number_Purchases

Total_Revenue = purchase_data["Price"].sum()
Total_Revenue

Purchase_Analysis = pd.DataFrame({"Number of Unique Items":[Item_Count], "Average Price":[Average_Price],
                                 "Number of Purchase":[Number_Purchases], "Total Revenue":[Total_Revenue]})
Purchase_Analysis["Average Price"] = Purchase_Analysis["Average Price"].map("${:}".format)
Purchase_Analysis["Total Revenue"] = Purchase_Analysis["Total Revenue"].map("${:}".format)

Purchase_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#.drop_duplicates can be added to end of brackets on first line for 1 line of large code
player_gender = purchase_data.loc[:,["SN", "Gender", "Age"]]
player_gender = player_gender.drop_duplicates()
player_gender
```

```python
#.value_counts can be applied to any column and will give value of column
player_data = player_gender["Gender"].value_counts()
player_data
```

```python
percent_player = round((player_data / total_players), 4)
percent = percent_player.map("{:.2%}".format)

gender_demographics_df = pd.DataFrame({"Total Count":player_data, "Percentage of Players":percent})
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data["Gender"].value_counts()

average_purchasep = purchase_data.groupby("Gender").mean()["Price"]

total_purchasev = purchase_data.groupby("Gender").sum()["Price"]

average_totalppp = total_purchasev / player_data


purchase_analysis_df = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": average_purchasep, 
                                    "Total Purchase Value": total_purchasev, 
                                    "Avg Total Purchase per Person": average_totalppp}) 

purchase_analysis_df = purchase_analysis_df.style.format({
    "Average Purchase Price": "${:.2f}",
    "Total Purchase Value": "${:.2f}",
    "Avg Total Purchase per Person": "${:.2f}"
})

purchase_analysis_df                                    
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40<"]
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]

player_gender["Age Range"] = pd.cut(player_gender["Age"], bins, labels=group_names, include_lowest=True)
player_gender

gender_count = player_gender.groupby("Age Range").count()["SN"]
percentage_count = gender_count / total_players
percentage_count

df_percent_players = pd.DataFrame({"Total Count":gender_count,
                                   "Percentage of Players":percentage_count}) 

df_percent_players['Percentage of Players'] = df_percent_players['Percentage of Players'].map("{:,.2%}".format)

df_percent_players
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
purchase_data

purchase_count = purchase_data.groupby("Age Range").count()["Purchase ID"]
purchase_count
average_purchasep = purchase_data.groupby("Age Range").mean()["Price"]
total_purchasev = purchase_data.groupby("Age Range").sum()["Price"]
average_totalppp = total_purchasev / gender_count

df_average_totalppp = pd.DataFrame({"Purchase Count":purchase_count,
                                   "Average Purchase Price":average_purchasep,
                                    "Total Purchase Value":total_purchasev,
                                    "Avg Total Purchase per Person":average_totalppp})

df_average_totalppp['Average Purchase Price'] = df_average_totalppp['Average Purchase Price'].map("${:,.2f}".format)
df_average_totalppp['Avg Total Purchase per Person'] = df_average_totalppp['Avg Total Purchase per Person'].map("${:,.2f}".format)
df_average_totalppp['Total Purchase Value'] = df_average_totalppp['Total Purchase Value'].map("${:,.2f}".format)

df_average_totalppp
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_count = purchase_data.groupby(["SN"]).count()["Purchase ID"]
spender_average_purchase_price = purchase_data.groupby(["SN"]).mean()["Price"]
spender_total_purchase_price = purchase_data.groupby(["SN"]).sum()["Price"]

top_spenders_df = pd.DataFrame({"Purchase Count":spender_count,
                                   "Average Purchase Price":spender_average_purchase_price, 
                                    "Total Purchase Value":spender_total_purchase_price})

top_spenders_df = top_spenders_df.sort_values("Total Purchase Value", ascending = False) 

top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map('${:,.2f}'.format)

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = purchase_data[["Item ID","Item Name","Price"]]

# group_item = item_df.groupby(["Item ID", "Item Name"]).count()["Item ID"]
# average_item = item_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
# sum_item = item_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

item_df = item_df.groupby(['Item ID','Item Name']).agg(
    item_count = pd.NamedAgg(column='Item ID', aggfunc='count'),
    average_item_price = pd.NamedAgg(column='Price', aggfunc='mean'),
    item_total_price = pd.NamedAgg(column='Price', aggfunc='sum')
)

item_df = item_df.sort_values("item_count", ascending = False) 
item_df.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]

item_df.head().style.format({
    "Item Price": "${:.2f}",
    "Total Purchase Value": "${:.2f}"
})
```
